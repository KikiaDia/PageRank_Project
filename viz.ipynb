```python
import matplotlib.pyplot as plt
import numpy as np

# Données
workers = [2, 4]  # Nombre de nœuds
dataframe_sans = [48.17, 32.50]  # DataFrame sans partition
dataframe_avec = [32.43, 20.13]  # DataFrame avec partition
rdd_sans = [45.55, 24.37]        # RDD sans partition

width = 0.1  # Largeur des barres

# Positions des barres
x = np.arange(len(workers))

# Création du graphique
fig, ax = plt.subplots(figsize=(10, 6))

# Création des barres
rects1 = ax.bar(x - width, dataframe_sans, width, label='DataFrame sans partition', color='royalblue')
rects2 = ax.bar(x, dataframe_avec, width, label='DataFrame avec partition', color='crimson')
rects3 = ax.bar(x + width, rdd_sans, width, label='RDD sans partition', color='orange')

# Personnalisation du graphique
ax.set_ylabel('Durée (minutes)')
ax.set_xlabel('Nombre de nœuds')
ax.set_title('Performance des différentes implémentations selon le nombre de nœuds')
ax.set_xticks(x)
ax.set_xticklabels(workers)
ax.legend()

# Ajout d'une grille horizontale
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Fonction pour ajouter les valeurs sur les barres
def autolabel(rects):
   for rect in rects:
       height = rect.get_height()
       ax.annotate(f'{height:.2f}',
                  xy=(rect.get_x() + rect.get_width() / 2, height),
                  xytext=(0, 3),  # 3 points de décalage vertical
                  textcoords="offset points",
                  ha='center', va='bottom')

# Ajout des labels sur les barres
autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# Ajustement de l'échelle Y pour avoir un peu d'espace au-dessus des barres
plt.ylim(0, max(max(dataframe_sans), max(dataframe_avec), max(rdd_sans)) * 1.1)

# Ajustement du layout
plt.tight_layout()

# Sauvegarde et affichage
plt.savefig('pagerank_performances_comparison.png', dpi=300, bbox_inches='tight')
plt.show()
```
